```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
c = (['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8'])
for i in c:
    train_x[i]=le.fit_transform(train_x[i])
    test_y[i]=le.fit_transform(test_y[i])
```

```python
# If for any column(s), the variance is equal to zero, then you need to remove those variable(s).

display(train_x.loc[:,train_x.var()==0].head())
display(test_y.loc[:,test_y.var()==0].head())
```

```python
train_xv = train_x.loc[:,train_x.var()!=0]
```

```python
test_yv = test_y.loc[:,test_y.var()!=0]
```

```python
train_yv = train_x['y']
train_xv = train_x.drop(['ID','y'],axis=1)
```

```python
test_yv = test_y.drop(['ID'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
```

```python
train_xvs = ss.fit_transform(train_x)
test_yvs = ss.fit_transform(test_y)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95,random_state=100)
```

```python
train_xvs_pca = pca.fit_transform(train_xvs)
test_yvs_pca = pca.fit_transform(test_yvs)
```

```python
print('Train Shape',train_x.shape)
print('Test Shape',test_y.shape)
print('Train after variance Shape',train_xv.shape)
print('Test after variance Shape',test_yv.shape)
print('Train after pca Shape{}'.format(train_xvs_pca.shape))
print('Test after pca Shape{}'.format(test_yvs_pca.shape))
```

```python
#XGBoost prediction
```

```python
#!pip install xgboost
from xgboost import XGBRegressor 
xgb = XGBRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
warnings.filterwarnings("ignore",category= DeprecationWarning)
warnings.filterwarnings("ignore", category= FutureWarning)
param_grid ={'n_estimators':[20,40,60,80],'learning_rate':[0.1,0.01,0.001,0.0001],'max_depth':[2,4,6,8]}
xgb_tuned = GridSearchCV(XGBRegressor(),param_grid=param_grid ,refit=True)
# fit the training data to model
xgb_tuned.fit(train_xv ,train_yv)
```

```python
print(xgb_tuned.best_params_)
```

```python
# predict 
xgb_tuned_predict = xgb_tuned.predict(test_yv)
```

```python
from sklearn.metrics import r2_score ,mean_absolute_error
```

```python
# check for model r2_score
r2_score_xbg_tuned =r2_score(train_yv,xgb_tuned_predict)
display('r2-score with the default paramer {}'.format(r2_score_xbg_tuned))
print()
mse_xgb_tuned =mean_absolute_error(train_yv,xgb_tuned_predict)
display('mean absolute error with the default paramer {}'.format(mse_xgb_tuned))
```
